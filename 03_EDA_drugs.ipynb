# Cuting edge (high sensitivity and low sensitivity) analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# IC50 plotting function for datasets with IC50 column
def load_and_log_ic50(filepath):
    df = pd.read_csv(filepath)
    if "IC50" in df.columns:
        ic50 = df["IC50"].replace([np.inf, -np.inf], np.nan).dropna()
        ic50 = ic50[ic50 > 0]
        return np.log(ic50)
    return None

# Define paths
dataset_paths = {
    "DrugSens": "/content/drive/MyDrive/VISTEC-dataset/datasets/sensitivity/DrugSens.csv",
    "DrugSens_withnull": "/content/drive/MyDrive/VISTEC-dataset/datasets/sensitivity/DrugSens_withnull.csv",
    "DrugSens_Train_regr_stack": "/content/drive/MyDrive/VISTEC-dataset/datasets/sensitivity/stack/regr/DrugSens-Train.csv",
    "DrugSens_Test_regr_stack": "/content/drive/MyDrive/VISTEC-dataset/datasets/sensitivity/stack/regr/DrugSens-Test.csv",
}

# Load and collect log(IC50) data
log_ic50_data = {}
for name, path in dataset_paths.items():
    try:
        log_vals = load_and_log_ic50(path)
        if log_vals is not None:
            log_ic50_data[name] = log_vals
    except Exception as e:
        print(f"Could not load {name}: {e}")

# Plot
if log_ic50_data:
    rows = int(np.ceil(len(log_ic50_data) / 2))
    fig, axs = plt.subplots(nrows=rows, ncols=2, figsize=(18, rows * 4))
    axs = axs.flatten()

    for i, (name, log_vals) in enumerate(log_ic50_data.items()):
        axs[i].hist(log_vals, bins=40, color='skyblue', edgecolor='black')
        axs[i].axvline(x=1.5, color='red', linestyle='--', label='cutoff ln(IC50)=1.5')
        axs[i].set_title(name)
        axs[i].set_xlabel("ln(IC50)")
        axs[i].set_ylabel("Frequency")
        axs[i].legend()

    for j in range(i + 1, len(axs)):
        fig.delaxes(axs[j])

    plt.tight_layout()
    plt.suptitle("Log(IC50) Distributions", fontsize=18, y=1.02)
    plt.show()
else:
    print("No datasets with IC50 values were successfully loaded.")
```

# Cleaning

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# 1: Load expression & sensitivity data
gene_exp = pd.read_csv('/content/drive/MyDrive/VISTEC-dataset/datasets/features/GeneExp.csv')
drug_sens = pd.read_csv('/content/drive/MyDrive/VISTEC-dataset/datasets/sensitivity/stack/regr/DrugSens-Train.csv')

# 2: Clean gene expression
# Drop columns with too many missing values (e.g., >30%)
gene_exp_clean = gene_exp.dropna(thresh=0.7 * gene_exp.shape[0], axis=1)

# Fill remaining missing values with column mean
gene_exp_clean = gene_exp_clean.fillna(gene_exp_clean.mean(numeric_only=True))

# 3: Standard scale gene features (except ID column)
gene_features = gene_exp_clean.drop(columns=['Unnamed: 0'])
scaler = StandardScaler()
gene_scaled = scaler.fit_transform(gene_features)
gene_scaled_df = pd.DataFrame(gene_scaled, columns=gene_features.columns)
gene_scaled_df.insert(0, 'cell_line', gene_exp_clean['Unnamed: 0'])

# 4: Prepare sensitivity data
# Remove rows with missing or invalid IC50
drug_sens = drug_sens[drug_sens['IC50'].notna() & (drug_sens['IC50'] > 0)]

# Add sensitivity label: 1 = high sensitivity, 0 = low sensitivity
drug_sens['sensitivity_label'] = (np.log(drug_sens['IC50']) <= 1.5).astype(int)

# 5: Merge gene expression with sensitivity data on cell line
merged = pd.merge(
    gene_scaled_df,
    drug_sens[['cell_line', 'gdsc_name', 'IC50', 'sensitivity_label', 'cancer_type']],
    on='cell_line',
    how='inner'
)

# 6: Final preview
print("Final dataset shape:", merged.shape)
print("Columns:", merged.columns.tolist())
print("Missing values:", merged.isnull().sum().sum())
print("Cancer types distribution:\n", merged['cancer_type'].value_counts())
print("Sensitivity label distribution:\n", merged['sensitivity_label'].value_counts())
```

```python
print("\n" + "="*50)
print("GENE CORRELATION NETWORK ANALYSIS")
print("="*50)

# Analyze correlations between top variable genes
top_corr_genes = merged[gene_cols].var().nlargest(30).index
corr_matrix = merged[top_corr_genes].corr()

print(f"Analyzing correlation network of top {len(top_corr_genes)} variable genes...")

# Find highly correlated gene pairs
high_corr_pairs = []
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        corr_val = corr_matrix.iloc[i, j]
        if abs(corr_val) > 0.7:  # High correlation threshold
            high_corr_pairs.append((corr_matrix.columns[i], corr_matrix.columns[j], corr_val))

high_corr_pairs.sort(key=lambda x: abs(x[2]), reverse=True)

print(f"Found {len(high_corr_pairs)} highly correlated gene pairs (|r| > 0.7):")
for gene1, gene2, corr in high_corr_pairs[:10]:
    print(f"  {gene1} - {gene2}: r = {corr:.3f}")

# Visualization: Correlation network
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

# Correlation heatmap
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=False, cmap='RdBu_r', center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .8}, ax=ax1)
ax1.set_title('Gene-Gene Correlation Network')

# Distribution of correlation values
corr_values = corr_matrix.values[np.triu_indices_from(corr_matrix.values, k=1)]
ax2.hist(corr_values, bins=50, alpha=0.7, color='purple', edgecolor='black')
ax2.axvline(x=0.7, color='red', linestyle='--', label='High correlation threshold')
ax2.axvline(x=-0.7, color='red', linestyle='--')
ax2.axvline(x=0, color='gray', linestyle='-', alpha=0.5)
ax2.set_xlabel('Correlation Coefficient')
ax2.set_ylabel('Frequency')
ax2.set_title('Distribution of Gene-Gene Correlations')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# 1: Define gene columns
gene_cols = [col for col in merged.columns if col not in ['cell_line', 'gdsc_name', 'IC50', 'sensitivity_label', 'cancer_type']]
gene_stats = merged[gene_cols].agg(['mean', 'std']).T
gene_stats['cv'] = gene_stats['std'] / gene_stats['mean'].replace(0, np.nan)

# 2: IQR Helper Function
def get_iqr_bounds(series, factor=1.5):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    return Q1 - factor * IQR, Q3 + factor * IQR

std_lower, std_upper = get_iqr_bounds(gene_stats['std'])

# 3: Identify Outlier Genes
zero_var_genes = gene_stats[gene_stats['std'] == 0]
high_mean_genes = gene_stats[gene_stats['mean'] > gene_stats['mean'].quantile(0.95)]
extreme_mean_genes = gene_stats[gene_stats['mean'] > gene_stats['mean'].quantile(0.99)]
outlier_std_genes = gene_stats[(gene_stats['std'] < std_lower) | (gene_stats['std'] > std_upper)]
high_cv_genes = gene_stats[gene_stats['cv'] > gene_stats['cv'].quantile(0.95)]

# 4: Visualize Distributions
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(gene_stats['mean'], bins=30, kde=True)
plt.title('Distribution of Gene Mean Values')
plt.xlabel('Mean')

plt.subplot(1, 2, 2)
sns.histplot(gene_stats['std'], bins=30, kde=True)
plt.title('Distribution of Gene Standard Deviation')
plt.xlabel('Standard Deviation')
plt.tight_layout()
plt.show()

# 5: QC Summary & Scatter
print("\n=== QUALITY CONTROL SUMMARY ===")
print(f"Total genes analyzed: {len(gene_stats)}")
print(f"Genes with zero variance: {len(zero_var_genes)}")
print(f"Highly expressed genes (>95th percentile): {len(high_mean_genes)}")
print(f"Outlier genes (by std): {len(outlier_std_genes)}")
print(f"Outlier percentage: {len(outlier_std_genes)/len(gene_stats)*100:.1f}%")
print(f"\nOutlier STD bounds: lower={std_lower:.3f}, upper={std_upper:.3f}")
print(f"Mean-Std correlation: {np.corrcoef(gene_stats['mean'], gene_stats['std'])[0,1]:.3f}")

# 6: Highlight Key Genes
print("\n=== TOP GENE HIGHLIGHTS ===")
print("Top 5 highest mean genes:")
print(high_mean_genes.nlargest(5, 'mean')[['mean', 'std']])

print("\nTop 5 most variable (std) genes:")
print(outlier_std_genes.nlargest(5, 'std')[['mean', 'std']])

print("\nTop 5 highest CV genes (potentially noisy):")
print(high_cv_genes.nlargest(5, 'cv')[['mean', 'std', 'cv']])

# 7: Advanced Plots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
fig.suptitle('Gene Expression Quality Control Analysis', fontsize=14, fontweight='bold')

# Histogram with outlier bounds
axes[0].hist(gene_stats['std'], bins=50, color='lightcoral', edgecolor='black')
axes[0].axvline(std_lower, color='red', linestyle='--', label=f'Lower bound: {std_lower:.2f}')
axes[0].axvline(std_upper, color='red', linestyle='--', label=f'Upper bound: {std_upper:.2f}')
axes[0].set_title('Gene Standard Deviation')
axes[0].set_xlabel('Std Dev')
axes[0].set_ylabel('Frequency')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Mean vs Std Dev scatter
axes[1].scatter(gene_stats['mean'], gene_stats['std'], alpha=0.6, s=20, color='steelblue')
axes[1].set_xlabel('Mean Expression')
axes[1].set_ylabel('Standard Deviation')
axes[1].set_title('Mean vs Std Dev')
z = np.polyfit(gene_stats['mean'], gene_stats['std'], 1)
p = np.poly1d(z)
axes[1].plot(gene_stats['mean'], p(gene_stats['mean']), "r--", label=f'Trend line (R²={np.corrcoef(gene_stats["mean"], gene_stats["std"])[0,1]**2:.3f})')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
